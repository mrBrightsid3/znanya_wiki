```python
import pandas as pd
import fasttext
```

```python
df = pd.read_excel('doc_comment_summary.xlsx', header=None)
df.columns = ['text', 'label']
df['text'] = df['text'].str.lower().str.split()
df = df.dropna()
df.info()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('russian')
df['text_without_stopwords'] = df['text'].apply(lambda x: [word for word in x if word not in stop_words])
df
```

```python
df['text_without_stopwords'] = df['text_without_stopwords'].apply(lambda x: ' '.join(x))
df
```

```python
# Сохранение текста в файл
df['text_without_stopwords'].to_csv('texts.txt', index=False, header=False)

# Обучение модели FastText
model = fasttext.train_unsupervised('texts.txt')
```

```python
embeddings = df['text_without_stopwords'].apply(lambda x: model.get_sentence_vector(x))
embeddings
```

```python
from sklearn.model_selection import train_test_split
X = embeddings.tolist()
y = df['label'].to_list()

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='sigmoid')
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
